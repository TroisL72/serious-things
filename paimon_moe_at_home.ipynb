```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_excel("paimonmoe_wish_history.xlsx", sheet_name=0)
df1.head()
```

```python
df2 = pd.read_excel("paimonmoe_wish_history.xlsx", sheet_name=1)
df2.head()
```

```python
df3 = pd.read_excel("paimonmoe_wish_history.xlsx", sheet_name=2)
df3.head()
```

```python
df = pd.concat([df1, df2, df3])
df.drop('Part', axis=1, inplace=True)
df.head()
```

```python
type = np.array(df['Type'])
name = np.array(df['Name'])
stars = np.array([df['Stars']])
pity = np.array(df['Pity'])
roll = np.array(df['#Roll'])
```

```python
name_stars = df[['Name', 'Stars']]
five_stars = []
for index, row in name_stars.iterrows():
    if row['Stars'] == 5:
        five_stars.append(row['Name'])

print(five_stars)
```

```python
pity_char = df[['Pity', 'Name', 'Stars']]
pity_5stars_counts = []
for index, row in pity_char.iterrows():
    if row['Stars'] == 5:
        print(f"Pity: {row['Pity']}, Character: {row['Name']}.")
        pity_5stars_counts.append(row['Pity'])
print('-'*100)
average_pity = np.mean(pity_char[pity_char['Stars'] == 5]['Pity'])
print(f'Average Pity: {int(average_pity)}')
```

```python
pity_char = df[['Pity', 'Name', 'Stars']]
pity_char_5stars = pity_char[pity_char['Stars'] == 5]

five_stars_counts = pity_char_5stars['Name'].value_counts()
occurrences = {}
for index, row in pity_char_5stars.iterrows():
    char_name = row['Name']
    if char_name in occurrences:
        occurrences[char_name] += 1
    else:
        occurrences[char_name] = 1

for char_name, count in occurrences.items():
    print(f"Character: {char_name}, Constellations: C{count - 1}")
```

```python
num_5_stars = len(pity_5stars_counts)
five_stars_shows = []
for i in range(num_5_stars):
    five_stars_shows.append(5)

fig, ax = plt.subplots()
ax.scatter(five_stars_shows, pity_5stars_counts)
ax.xaxis.set_visible(False)
plt.title('5 Stars Character Pity Plot')
plt.ylabel('Pity')
plt.xlabel('Characters')
plt.show()
```

```python
pity_char = df[['Pity', 'Name', 'Stars', 'Type']]
pity_4stars_counts = []
for index, row in pity_char.iterrows():
    if row['Stars'] == 4:
        print(f"Pity: {row['Pity']}, Name: {row['Name']}.")
        pity_4stars_counts.append(row['Pity'])
print()
average_pity = np.mean(pity_char[pity_char['Stars'] == 4]['Pity'])
print(f'Average Pity: {int(average_pity)}')
```

```python
pity_char_4stars = df[(df['Stars'] == 4)]
four_stars_counts = pity_char_4stars['Name'].value_counts()
occurrences = {}
for index, row in pity_char_4stars.iterrows():
    char_name = row['Name']
    if char_name in occurrences:
        occurrences[char_name] += 1
    else:
        occurrences[char_name] = 1

for char_name, count in occurrences.items():
    if count >=6:
        count = 7
    print(f"Name: {char_name}, Constellations: C{count - 1}")
```

```python
num_4_stars = len(pity_4stars_counts)
four_stars_shows = []
for i in range(num_4_stars):
    four_stars_shows.append(4)

fig, ax = plt.subplots()
ax.scatter(four_stars_shows, pity_4stars_counts)
ax.xaxis.set_visible(False)
plt.title('4 Stars Character Pity Plot')
plt.ylabel('Pity')
plt.xlabel('Characters')
plt.show()
```
